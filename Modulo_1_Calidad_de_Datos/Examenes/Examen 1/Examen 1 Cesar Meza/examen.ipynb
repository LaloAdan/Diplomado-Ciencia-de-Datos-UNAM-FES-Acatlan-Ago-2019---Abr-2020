# Ingenierìa de Datos

```python
# 1. Construya una función a partir del conjunto de entrenamiento (train.csv) que
# genere al menos tres variables en función de las existentes, recibiendo como parámetro la tabla

# Promedio de la acciòn diaria
# Promedio Esperado (Màximo + Mìnimo)/2
# Promedio Ponderado (Promedio acciòn diaria + promedio esperado) /2

def ejercicio_1(tabla):

    tabla["Prom_diario"] = (tabla["Apertura"] + tabla["Cierre"])/2
    tabla["Prom_Esp"] = (tabla['Máximo'] + tabla['Mínimo'])/2
    tabla["Prom_Pond"] = (tabla["Prom_diario"] + tabla["Prom_Esp"])/2
    
    return tabla
ejercicio_1(train)
train.head(15)
```

```python
# 2. Usando la función generada previamente, aplique el mismo proceso a los datos de test
ejercicio_1(test)
test.head(10)
```

```python
# 3. Genere una función que genere las variables de: dı́a, mes y año a partir de la columna Fecha

def ejercicio_3(tabla):
    tabla['Fecha'] = pd.to_datetime(tabla['Fecha'])
    tabla["dia"] = tabla["Fecha"].dt.strftime("%d")
    tabla["mes"] = tabla["Fecha"].dt.strftime("%m")
    tabla["anio"] = tabla["Fecha"].dt.strftime("%Y")
    
    return tabla
ejercicio_3(test)
test.head(15)
```

```python
# 4. Genere una función que genere las variables de: dı́a, mes y año a partir de la columna Fecha

def ejercicio_4(tabla):
    Agrup_1 = tabla[["anio", "mes","Prom_diario"]].groupby(["anio", "mes"]).agg(["mean", "count"])
    Agrup_2 = tabla[["anio", "Quarter","Prom_diario"]].groupby(["anio", "Quarter"]).agg(["mean", "count"])
    
    return Agrup_1, Agrup_2
```

```python
# 5. Aplique la función al conjunto de test
ejercicio_4(test)
```

```python
def ks_var_continuas(data_1,data_2,var):
    stadistic_a, p_value_a = ks_2samp(data_1[var], data_2[var])
    # return var, stadistic_a, p_value_a
    return p_value_a
```

```python
Cat_var_cont = {}
for var_cont in ls_cont:
    aux = ks_var_continuas(train,test,var_cont)
    if aux >= .05: Cat_var_cont[var_cont] = [aux]
        
Cat_var_cont
```
